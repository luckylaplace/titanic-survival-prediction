# ðŸ’¡ 1. Ã–zellik MÃ¼hendisliÄŸi Fikirleri

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("../data/raw/train.csv")
test = pd.read_csv("../data/raw/test.csv")
```

```python
# 1. FamilySize
train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
test["FamilySize"] = test["SibSp"] + test["Parch"] + 1
```

```python
# 2. IsAlone
train['IsAlone'] = (train['FamilySize'] == 1).astype(int)
test['IsAlone'] = (test['FamilySize'] == 1).astype(int)
```

```python
# 3. Title Ã§Ä±kar
train['Title'] = train['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
test['Title'] = test['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
# Nadir title'larÄ± grupla
train['Title'] = train['Title'].replace(['Lady', 'Countess', 'Capt', 'Col',
                                         'Don', 'Dr', 'Major', 'Rev', 'Sir',
                                         'Jonkheer', 'Dona'], 'Rare')
train['Title'] = train['Title'].replace('Mlle', 'Miss')
train['Title'] = train['Title'].replace('Ms', 'Miss')
train['Title'] = train['Title'].replace('Mme', 'Mrs')
```

```python
# Test iÃ§in de aynÄ± iÅŸlem
test['Title'] = test['Title'].replace(['Lady', 'Countess', 'Capt', 'Col',
                                       'Don', 'Dr', 'Major', 'Rev', 'Sir',
                                       'Jonkheer', 'Dona'], 'Rare')
test['Title'] = test['Title'].replace('Mlle', 'Miss')
test['Title'] = test['Title'].replace('Ms', 'Miss')
test['Title'] = test['Title'].replace('Mme', 'Mrs')
```

```python
print("âœ… Yeni Ã¶zellikler oluÅŸturuldu")
```

```python
# 4. Age eksiklerini doldur (Pclass + Sex kombinasyonuna gÃ¶re median)
for dataset in [train, test]:
    for pclass in [1, 2, 3]:
        for sex in ['male', 'female']:
            median_age = dataset[(dataset['Pclass']==pclass) & 
                                (dataset['Sex']==sex)]['Age'].median()
            dataset.loc[(dataset['Age'].isnull()) & 
                       (dataset['Pclass']==pclass) & 
                       (dataset['Sex']==sex), 'Age'] = median_age

print("âœ… Age eksikleri dolduruldu")
```

```python
# 5. Embarked eksiklerini doldur (en sÄ±k olan)
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])

# 6. Test'teki Fare eksiklerini doldur
test['Fare'] = test['Fare'].fillna(test['Fare'].median())

print("âœ… TÃ¼m eksikler dolduruldu")
```

```python
# Kaydet
train.to_csv('../data/processed/train_cleaned.csv', index=False)
test.to_csv('../data/processed/test_cleaned.csv', index=False)

print("âœ… TemizlenmiÅŸ veriler kaydedildi")
```
